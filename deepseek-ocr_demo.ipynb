```python
from transformers import AutoModel, AutoTokenizer
import torch
model_name = 'deepseek-ai/DeepSeek-OCR'

tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
model = AutoModel.from_pretrained(model_name,trust_remote_code=True, torch_dtype=torch.bfloat16, device_map="auto",use_safetensors=True)
model = model.eval()
print("model loaded!!")
```

```python
prompt = (
    "<image>\n"
    "Perform OCR on this receipt. "
    "Extract visible number and text accurately," 
)

image_file = "/kaggle/input/testingkwit/kwitansi_1.jpg"
output_path = "/kaggle/working/ocr_results"

!mkdir -p /kaggle/working/ocr_results

res = model.infer(
    tokenizer,
    prompt=prompt,
    image_file=image_file,
    output_path=output_path,
    base_size=1024,
    image_size=640,
    crop_mode=True,
    save_results=True,
    test_compress=True
)
```

```python
import re
import json

# =====================================
# INPUT FILE OCR
# =====================================
OCR_FILE = "/kaggle/working/ocr_results/result.mmd"

# =====================================
# LOAD OCR TEXT
# =====================================
with open(OCR_FILE, "r", encoding="utf-8") as f:
    ocr_text = f.read()

# =====================================
# HELPER FUNCTIONS
# =====================================
def clean_text(text):
    text = text.replace('\n', ' ')
    text = re.sub(r'\s+', ' ', text)
    return text.strip()

def clean_money(value):
    return int(value.replace('.', '').replace(',', '').strip())

def normalize_date(d, m, y):
    if len(y) == 2:
        y = "20" + y
    return f"{y}-{m.zfill(2)}-{d.zfill(2)}"

# =====================================
# PRE-CLEAN
# =====================================
text = clean_text(ocr_text)

# =====================================
# NER EXTRACTION
# =====================================

# 1. PAYMENT DATE
payment_date = None
date_match = re.search(r'(\d{1,2})-(\d{1,2})-(\d{2,4})', text)
if date_match:
    d, m, y = date_match.groups()
    payment_date = normalize_date(d, m, y)

# 2. NAME
name_match = re.search(r'Nama\s*:\s*([A-Za-z ]+)', text)
student_name = name_match.group(1).strip() if name_match else None

# 3. CLASS
class_match = re.search(r'Kelas\s*:\s*([A-Za-z0-9]+)', text)
student_class = class_match.group(1).strip() if class_match else None

# 4. SUBJECT
subject_match = re.search(r'Subject\s*:\s*([A-Za-z /]+)', text)
subjects = []
if subject_match:
    subjects = [s.strip() for s in subject_match.group(1).split('/')]

# 5. TOTAL AMOUNT
total_match = re.search(r'TOTAL\s*:\s*Rp\.?\s*([\d\.]+)', text)
total_amount = clean_money(total_match.group(1)) if total_match else None

# 6. SPP AMOUNT (heuristic)
spp_amount = None
if total_amount:
    if total_amount in [420000, 470000]:
        spp_amount = total_amount

# =====================================
# FINAL STRUCTURED OUTPUT
# =====================================
ner_result = {
    "payment_date": payment_date,
    "student": {
        "name": student_name,
        "class": student_class,
        "subject": subjects
    },
    "payment": {
        "type": "SPP",
        "amount": spp_amount
    },
    "total_amount": total_amount
}

# =====================================
# OUTPUT
# =====================================
print("\n===== NER RESULT =====")
print(json.dumps(ner_result, indent=2, ensure_ascii=False))
```
